## 피클 파일 열기

```python
import pandas as pd
import pickle

p = open('products.pkl', 'rb')
p_df = pickle.load(p)
p.close()
```

## 옵션에서 괄호 및 괄호 내 텍스트 제거

```python
import re

p_df['옵션'] = p_df['옵션'].astype('str')

def cleanText(readData):
 
    # 괄호랑 그 내용 없애기
    text = re.sub(r'\([^)]*\)', '', readData)
    return text


p_df['옵션'] = p_df['옵션'].apply(cleanText)
p_df['옵션'] = p_df['옵션'].str.replace('[', '(').str.replace(']', ')')
p_df['옵션'] = p_df['옵션'].apply(cleanText)
```

## 상품명에서 특수문자 제거하기

```python
import re

p_df['상품명'] = p_df['상품명'].astype('str')

def cleanText(readData):
 
    # 괄호랑 그 내용 없애기
    text = re.sub(r'\([^)]*\)', '', readData)
    return text


p_df['상품명'] = p_df['상품명'].apply(cleanText)
p_df['상품명'] = p_df['상품명'].str.replace('[', '(').str.replace(']', ')')
p_df['상품명'] = p_df['상품명'].apply(cleanText)
```

## 상품명에서 공백과 특수문자(>) 제거

```python

p_df['상품명'] = p_df['상품명'].str.replace('>','')
p_df['상품명'] = p_df['상품명'].str.strip()
```

 ## 색상 칼럼 추가 

```python
p_df['색상'] =""
```

```python
import numpy as np
conditions = [
    (p_df['옵션'].notnull()  & p_df['옵션'].str.contains('색상선택 :')),
    (p_df['옵션'].notnull()  & p_df['옵션'].str.contains('색상 선택 :')),
    (p_df['옵션'].notnull()  & p_df['옵션'].str.contains('색상 :' )),
   
   ]

choices = [p_df['옵션'].str.split('색상선택 :',expand=True)[1],
            p_df['옵션'].str.split('색상 선택 :',expand=True)[1],
            p_df['옵션'].str.split('색상 :',expand=True)[1]
          
          ]

p_df['색상'] = np.select(conditions, choices, default="")
```

```python
p_df
```

## 쉼표  앞만 추출하기

```python

p_df['색상'] = p_df['색상'].str.split(',', expand=True)[0]
```

## 공백 제거

```python
p_df['색상'] = p_df['색상'].str.strip()
```

## 1개 2개 3개 4개 5개 6개 7개 없애기

```python
p_df['색상'] = p_df['색상'].str.replace('1개','')
p_df['색상'] = p_df['색상'].str.replace('2개','')
p_df['색상'] = p_df['색상'].str.replace('3개','')
p_df['색상'] = p_df['색상'].str.replace('4개','')
p_df['색상'] = p_df['색상'].str.replace('5개','')
p_df['색상'] = p_df['색상'].str.replace('6개','')
p_df['색상'] = p_df['색상'].str.replace('7개','')
```

## 공백 제거 한번 더

```python
p_df['색상'] = p_df['색상'].str.strip()
```
